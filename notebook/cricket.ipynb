```python
#chnage data types 
df['start_date']=pd.to_datetime(df['start_date'])
```

```python
#check null vaues
df.isnull().sum()
```

```python
df['wicket_type']=df['wicket_type'].fillna('No Wicket')
df['total_runs'] = df['runs_off_bat'] + df['extras']
```
